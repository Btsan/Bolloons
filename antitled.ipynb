```python
import win32gui
import pyautogui

def get_hwnd(name='BloonsTD6'):
    return win32gui.FindWindow(None, name)

def get_bloons_screen_cords():
    hwnd = get_hwnd('BloonsTD6')
    left, top, right, bot = win32gui.GetWindowRect(hwnd)
    return (left, top, right, bot)

def get_bloons_screen():
    left, top, right, bot = get_bloons_screen_cords()
    height = bot-top
    width = right-left
    img = pyautogui.screenshot(region=(left, top, width, height))
    return img
```

```python
hwnd = get_hwnd()
old_coords = get_bloons_screen_cords()
win32gui.MoveWindow(hwnd, 0, 0, 1616, 939, True)
img = get_bloons_screen()
win32gui.MoveWindow(hwnd, *old_coords, True)
```
